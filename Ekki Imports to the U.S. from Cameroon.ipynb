### The number of companies supplying the U.S. with greenheart from Guyana has fluctuated over the years, with a high in 2013, and few imports in recent years

```python
df.groupby('year')['GROSS WEIGHT (KG)'].sum()
```

```python
df.groupby('year')['GROSS WEIGHT (KG)'].sum().plot()
```

### The weight imported fluctuates, but has declined since around 2013. No shipments direct from Cameroon in 2020, 2021, 2022 or 2024

```python
df.groupby(['year','SHIPPER_CLEANED'])['GROSS WEIGHT (KG)'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Group and aggregate
df_grouped = df.groupby(['year', 'SHIPPER_CLEANED'])['GROSS WEIGHT (KG)'].sum().reset_index()
shippers = df_grouped['SHIPPER_CLEANED'].unique()

# Set number of subplots
n = len(shippers)
cols = 3
rows = -(-n // cols)  # ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4), sharex=True)
axes = axes.flatten()

for i, shipper in enumerate(shippers):
    shipper_df = df_grouped[df_grouped['SHIPPER_CLEANED'] == shipper]
    axes[i].bar(shipper_df['year'].astype(str), shipper_df['GROSS WEIGHT (KG)'])  # year as string for categorical x-axis
    axes[i].set_title(shipper)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Gross Weight (KG)')
    axes[i].grid(True, axis='y')

# Remove any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### New York (Newark and NYC) have imported over half of all imports to the U.S. of ekki

```python
df.groupby('US PORT')['GROSS WEIGHT (KG)'].sum().sort_values(ascending=False)
```

```python
# Total gross weight per port
port_weights = df.groupby('US PORT')['GROSS WEIGHT (KG)'].sum()

# Normalize by dividing by total and multiplying by 100 (for percentage)
port_weights_percent = (port_weights / port_weights.sum()) * 100

# Sort descending
port_weights_percent = port_weights_percent.sort_values(ascending=False)

port_weights_percent
```

### New York has imported 35% of the total weight of ekki imported directly from Cameroon, following Houston

```python
# Group and aggregate
df_grouped = df.groupby(['year', 'US PORT'])['GROSS WEIGHT (KG)'].sum().reset_index()
ports = df_grouped['US PORT'].unique()

# Set up subplot grid
n = len(ports)
cols = 3
rows = -(-n // cols)  # ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4), sharex=True)
axes = axes.flatten()

# Plot each port
for i, port in enumerate(ports):
    port_df = df_grouped[df_grouped['US PORT'] == port]
    axes[i].bar(port_df['year'].astype(str), port_df['GROSS WEIGHT (KG)'])
    axes[i].set_title(port)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Gross Weight (KG)')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(True, axis='y')

# Remove any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
