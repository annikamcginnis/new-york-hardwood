# ITTO Timber Trade Data Analysis

## Guyana

```python
import pandas as pd
```

```python
df = pd.read_csv('Itto_20250625_085340.csv')
```

```python
df.head()
```

```python
df['Superscript'].value_counts()
```

```python
df.shape
```

```python
df['Product'].value_counts()
```

```python
df['Flows'].value_counts()
```

```python
df[df['Flows'] == 'Exports Quantity'].groupby('Product')['Value OR Volume'].sum().sort_values(ascending=False)
```

```python
tropical_hardwood_categories = [
    "Sawnwood", "Sawnwood (NC)", "Sawnwood (NC.T.)",
    "Ind. roundwood", "Ind. roundwood (NC)", "Ind. roundwood (NC.T.)",
    "Veneer (NC.T.)", "Plywood (NC.T.)",
    "Veneer", "Plywood"  # assume tropical if from Guyana
]
```

```python
guyana_tropical_exports = df[df['Product'].isin(tropical_hardwood_categories)]
```

```python
guyana_tropical_exports.head()
```

```python
guyana_tropical_exports[guyana_tropical_exports['Flows'] == 'Exports Quantity'].groupby('Year')['Value OR Volume'].sum()
```

```python
guyana_tropical_exports[guyana_tropical_exports['Flows'] == 'Exports Quantity'].groupby('Year')['Value OR Volume'].sum().plot()
```

```python
guyana_tropical_exports[guyana_tropical_exports['Flows'] == 'Exports Value'].groupby('Year')['Value OR Volume'].sum().plot()
```

```python
greenheart_categories = [
    "Sawnwood", "Sawnwood (NC)", "Sawnwood (NC.T.)",
    "Ind. roundwood", "Ind. roundwood (NC)", "Ind. roundwood (NC.T.)"
]
```

```python
greenheart_df = df[df['Product'].isin(greenheart_categories)]
```

```python
greenheart_df.head()
```

```python
greenheart_df[greenheart_df['Flows'] == 'Exports Quantity'].groupby('Year')['Value OR Volume'].sum().plot()
```

```python
greenheart_df[greenheart_df['Flows'] == 'Exports Value'].groupby('Year')['Value OR Volume'].sum()
```

```python
greenheart_df[greenheart_df['Flows'] == 'Exports Value'].groupby('Year')['Value OR Volume'].sum().plot()
```
