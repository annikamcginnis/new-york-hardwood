### The weight imported fluctuates, but has very slightly declined since around 2008.

```python
df.groupby(['year','SHIPPER_CLEANED'])['GROSS WEIGHT (KG)'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Group and aggregate
df_grouped = df.groupby(['year', 'SHIPPER_CLEANED'])['GROSS WEIGHT (KG)'].sum().reset_index()
shippers = df_grouped['SHIPPER_CLEANED'].unique()

# Set number of subplots
n = len(shippers)
cols = 3
rows = -(-n // cols)  # ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4), sharex=True)
axes = axes.flatten()

for i, shipper in enumerate(shippers):
    shipper_df = df_grouped[df_grouped['SHIPPER_CLEANED'] == shipper]
    axes[i].bar(shipper_df['year'].astype(str), shipper_df['GROSS WEIGHT (KG)'])  # year as string for categorical x-axis
    axes[i].set_title(shipper)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Gross Weight (KG)')
    axes[i].grid(True, axis='y')

# Remove any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

- Companies that used to import a lot but have dropped off: Durable Wood Products (CITY SUPPLIER), Variety Woods & Greenheart (CITY SUPPLIER), Demerara Timbers, Bulkan Timber Works

- Used to import in some years but have dropped off: Barama Co. Ltd., Guyana Timber Products, Trade Linc (CITY SUPPLIER), Toolsie Persaud (CITY SUPPLIER), Guyana Furniture Manufacturing, Trans Amazon Exporters, Cummings Wood Products, Emerald Forest (CITY SUPPLIER), Interior Forest Industries Ltd., Timberwood Products (CITY SUPPLIER)

- Didn't used to import but now import a lot: Woodpecker Timbers (CITY SUPPLIER), Prestige Timbers (CITY SUPPLIER), Nazir Tropical Timbers (CITY SUPPLIER), Parika Sawmills, McVantage of Guyana (CITY SUPPLIER)
  
- Now import in some years: G.S. Forest Resources (CITY SUPPLIER), Superior Woods, Woods Direct (CITY SUPPLIER), Deoraj Singh Sawmill, Reginald Manbodh/GS Forest (CITY SUPPLIER), Evergreen Forest Products, Fores Woods Inc.

```python
# Group and aggregate
df_grouped = df.groupby(['year', 'US PORT'])['GROSS WEIGHT (KG)'].sum().reset_index()
ports = df_grouped['US PORT'].unique()

# Set up subplot grid
n = len(ports)
cols = 3
rows = -(-n // cols)  # ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4), sharex=True)
axes = axes.flatten()

# Plot each port
for i, port in enumerate(ports):
    port_df = df_grouped[df_grouped['US PORT'] == port]
    axes[i].bar(port_df['year'].astype(str), port_df['GROSS WEIGHT (KG)'])
    axes[i].set_title(port)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Gross Weight (KG)')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(True, axis='y')

# Remove any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### Newark saw an increase in 2023

### New York saw big increases in 2022 and 2019
