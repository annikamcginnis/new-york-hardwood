# Filtering Cameroon Forest Concessions Shapefiles

Data download link: https://data-minfof.opendata.arcgis.com/datasets/c0596c5af1204e56b23104551d7397c1_0/explore?location=5.597335%2C12.490611%2C6.74

```python
import geopandas as gpd
```

```python
# Load the shapefile (just provide the path to the .shp file)
gdf = gpd.read_file("Cameroon forest concessions shapefiles/Forêts_de_production.shp")
```

```python
gdf.head()
```

```python
gdf = gdf.to_crs(epsg=4326)
```

```python
target_ids = ['09-021', '09-024', '11-005', '09-022', '09-025', '00-003']
gdf_filtered = gdf[gdf['nom_foret'].isin(target_ids)]
```

```python
gdf_filtered
```

```python
gdf_filtered.to_file("wijma_former_concessions.shp", driver='ESRI Shapefile')
```

```python
gdf_filtered.to_file("wijma_former_concessions.geojson", driver="GeoJSON")
```
