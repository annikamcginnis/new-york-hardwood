```python
import pandas as pd
```

```python
df = pd.read_csv('Greenheart from Guyana to US 2.csv')
```

```python
df.head()
```

```python
df_long = df.melt(
    id_vars=["YEAR"],           # Keep YEAR as identifier
    var_name="Port",            # Name of new column for port names
    value_name="imports"        # Name of new column for values
)
```

```python
df_long
```

```python
df_long.dtypes
```

```python
df_long = df_long.dropna(subset=["YEAR"])
```

```python
df_long = df_long[df_long["Port"].str.strip() != ""]
```

```python
df_long.head(40)
```

```python
df_long.to_csv('guyana_imports_long.csv',index=False)
```
