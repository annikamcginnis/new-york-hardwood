## Fuzzy matching the names of shippers to group together the same companies 

```python
unique_shippers = df['SHIPPER'].unique()
```

```python
len(unique_shippers)
```

```python
pip install 'thefuzz[speedup]'
```

```python
from thefuzz import fuzz

threshold = 85
groups = []
used = set()

for name in unique_shippers:
    if name in used:
        continue
    group = [name]
    used.add(name)
    for other in unique_shippers:
        if other not in used and fuzz.token_set_ratio(name, other) >= threshold:
            group.append(other)
            used.add(other)
    groups.append(group)
```

```python
name_to_rep = {}
for group in groups:
    representative = group[0]
    for name in group:
        name_to_rep[name] = representative
```

```python
df['SHIPPER_CLEANED'] = df['SHIPPER'].map(name_to_rep)
```

## Extracting out the tree species from the description column and creating new columns per species

```python
descriptions = df['PRODUCT DESCRIPTION'].unique()
```

```python
len(descriptions)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

API_KEY = os.getenv('API_KEY')
```

```python
from openai import OpenAI
import math

client = OpenAI(
    api_key=API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

# Known tree species list
known_species_list = [
    'baromalli', 'birch', 'kakaralli', 'bulletwood', 'burada', 'cara euc', 'cara pino',
    'cedar', 'greenheart', 'crabwood', 'dakama', 'darina', 'wamaradan', 'kabukalli',
    'limonaballi', 'locust', 'mora', 'purpleheart', 'pine', 'cuiabano pine', 'shibadan',
    'tatabu', 'tauroniro', 'virola', 'wallaba', 'wamara', 'washiba', 'yarula'
]

known_species_str = ", ".join(known_species_list)
known_species_lower = set(s.lower() for s in known_species_list)

all_new_species = set()

def extract_species_batch(batch):
    # Join the batch into a formatted bullet list
    joined_descriptions = "\n".join([f"- {desc}" for desc in batch])

    prompt = f"""
Below are several shipment descriptions that may contain names of tree species:

{joined_descriptions}

You already know about the following tree species:
{known_species_str}

Please return a comma-separated list of any additional tree species mentioned that are NOT in the known list. Only return the names of new species. If none, respond "None".
"""

    try:
        completion = client.chat.completions.create(
            model="gemini-2.0-flash",
            messages=[
                {"role": "system", "content": "You are an expert botanist identifying tree species in shipping descriptions."},
                {"role": "user", "content": prompt},
            ],
        )
        return completion.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error: {e}")
        return "Error"

# Split descriptions into batches
batch_size = 200
num_batches = math.ceil(len(descriptions) / batch_size)

for i in range(num_batches):
    batch = descriptions[i * batch_size : (i + 1) * batch_size]
    response = extract_species_batch(batch)

    if response and response.lower() not in {"none", "error"}:
        new_species = [s.strip().lower() for s in response.split(",")]
        for species in new_species:
            if species and species not in known_species_lower:
                all_new_species.add(species)

    print(f"Processed batch {i + 1}/{num_batches}")

print("\n=== Discovered Unknown Species ===")
print(sorted(all_new_species))
```

```python
# Combine and deduplicate
all_species = sorted(set(s.lower() for s in known_species_list) | set(all_new_species))
```

```python
for species in all_species:
    # Create a case-insensitive regex pattern
    pattern = rf"\b{re.escape(species)}\b"

    # Apply boolean mask: TRUE if species name is present in the description
    df[species] = df['PRODUCT DESCRIPTION'].str.contains(pattern, case=False, regex=True)
```

## How many different wood species did each company import per year?

```python
# Filter to relevant columns plus year and company
df_subset = df[['YEAR', 'IMPORTER_CLEANED'] + species_cols]

# Group by year and company, then sum the boolean columns (True = 1, False = 0)
species_counts = df_subset.groupby(['YEAR', 'IMPORTER_CLEANED'])[species_cols].sum()

# For each year and company, count how many species have at least 1 import (i.e. sum > 0)
species_counts['num_species_imported'] = (species_counts > 0).sum(axis=1)

# Optional: reset index for easier viewing
species_counts = species_counts.reset_index()

# Show just the count per company per year
result = species_counts[['YEAR', 'IMPORTER_CLEANED', 'num_species_imported']]

result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'result' is from previous step with columns YEAR, IMPORTER_CLEANED, num_species_imported

companies = (
    result.groupby('IMPORTER_CLEANED')['num_species_imported']
    .sum()
    .sort_values(ascending=False)
    .index
)

plot_data = result[result['IMPORTER_CLEANED'].isin(companies)]

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=plot_data,
    x='YEAR',
    y='num_species_imported',
    hue='IMPORTER_CLEANED',
    marker='o'
)

plt.title('Number of Different Wood Species Imported Per Year by Companies')
plt.ylabel('Number of Species Imported')
plt.xlabel('Year')
plt.legend(title='Importer', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
species = pd.read_csv('species_vulnerable_status.csv')
```

```python
df_melted = df.melt(
    id_vars=['YEAR', 'GROSS WEIGHT (KG)'],  # include weight column
    value_vars=species_cols,
    var_name='Species',
    value_name='Imported'
)
```

```python
df_melted.head()
```

```python
# Create lowercase versions of the species columns for merging
df_melted['Species_lower'] = df_melted['Species'].str.lower()
species['Species Name_lower'] = species['Species Name'].str.lower()

# Perform the merge using the lowercase versions
df_threatened = df_melted.merge(
    species[
        ['Species Name_lower', 'Threatened',
         'CITES Status: https://cites.org/sites/default/files/eng/app/2025/E-Appendices-2025-02-07.pdf',
         'IUCN Status', 'IUCN Population Trend']
    ],
    left_on='Species_lower',
    right_on='Species Name_lower',
    how='left'
)

# Optionally drop the lowercase helper columns
df_threatened.drop(['Species_lower', 'Species Name_lower'], axis=1, inplace=True)
```

```python
df_threatened.head()
```

```python
# Step 1: Group by YEAR and Threatened, summing the gross weight
threatened_weight_per_year = (
    df_threatened.groupby(['YEAR', 'Threatened'])['GROSS WEIGHT (KG)']
    .sum()
    .unstack(fill_value=0)
    .sort_index()
)

# Step 2: Plot
threatened_weight_per_year.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    colormap='Set2'
)

plt.title('Total Gross Weight of Wood Imports by Threatened Status per Year')
plt.xlabel('Year')
plt.ylabel('Gross Weight (KG)')
plt.legend(title='Threatened Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
