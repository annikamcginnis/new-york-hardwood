# Combining Cameroon Government Timber Export Data

```python
import pandas as pd
```

```python
companies_logs = pd.read_csv('cameroon_gov_data/Cameroon export reports by the Ministry of Forestry and Wildlife - companies_logs_transposed.csv')
```

```python
companies_logs.head()
```

```python
companies_sawnwood = pd.read_csv('cameroon_gov_data/Cameroon export reports by the Ministry of Forestry and Wildlife - companies_sawnwood_transposed.csv')
```

```python
companies_sawnwood
```

```python
countries_logs = pd.read_csv('cameroon_gov_data/Cameroon export reports by the Ministry of Forestry and Wildlife - countries_logs_transposed.csv')
```

```python
countries_logs
```

```python
countries_sawnwood = pd.read_csv('cameroon_gov_data/Cameroon export reports by the Ministry of Forestry and Wildlife - countries_sawnwood_transposed.csv')
```

```python
countries_sawnwood.head()
```

```python
countries_sawnwood['Year'] = countries_sawnwood['Destination']
```

```python
countries_sawnwood
```

```python
countries_sawnwood = countries_sawnwood.drop(columns='Destination')
```

```python
countries_sawnwood
```

```python
def clean_numeric(df):
    df = df.copy()
    for col in df.columns:
        if col != 'Year':
            # Remove commas and convert to float
            df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')
    return df
```

```python
companies_logs_clean = clean_numeric(companies_logs).set_index('Year')
companies_sawnwood_clean = clean_numeric(companies_sawnwood).set_index('Year')
countries_logs_clean = clean_numeric(countries_logs).set_index('Year')
countries_sawnwood_clean = clean_numeric(countries_sawnwood).set_index('Year')
```

```python
# Identify all numeric columns except 'Year'
numeric_cols = companies_sawnwood_clean.select_dtypes(include='number').columns.difference(['Year'])

# Multiply those numeric columns by 1.82
companies_sawnwood_clean[numeric_cols] = companies_sawnwood_clean[numeric_cols] * 1.82
```

```python
companies_sawnwood_clean.head()
```

```python
# Identify all numeric columns except 'Year'
numeric_cols = countries_sawnwood_clean.select_dtypes(include='number').columns.difference(['Year'])

# Multiply those numeric columns by 1.82
countries_sawnwood_clean[numeric_cols] = countries_sawnwood_clean[numeric_cols] * 1.82
```

```python
countries_sawnwood_clean
```

```python
countries_sawnwood_clean.dtypes
```

```python
# Convert all non-Year columns to numeric, coercing errors to NaN
def clean_numeric(df):
    df = df.copy()
    for col in df.columns:
        if col != 'Year':
            df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

# Apply conversion
logs_clean = clean_numeric(companies_logs_clean)
sawnwood_clean = clean_numeric(companies_sawnwood_clean)

# Add them together
combined_companies = logs_clean.add(sawnwood_clean, fill_value=0)

# Restore 'Year' as a column
combined_companies = combined_companies.reset_index()
```

```python
combined_companies.tail()
```

```python
combined_companies.to_csv('combined_companies.csv', index=False)
```
